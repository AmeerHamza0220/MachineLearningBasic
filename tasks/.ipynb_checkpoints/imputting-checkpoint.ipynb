```python
%%time
import seaborn as sns
features = list(data.columns)
features_1, features_2, features_3, features_4 = [], [], [], []
F = [[], [], [], [], []]
for feature in features:
    for i in [1, 2, 3, 4]:
        if feature.split('_')[1] == str(i):
            F[i].append(feature)
df = [[], [], [], [], []]

fig, axs = plt.subplots(nrows=4, ncols=1, figsize=(18, 30))

for i in [1, 2, 3, 4]:
    df[i] = data[F[i]]
    corr = df[i].corr()
    sns.heatmap(corr, ax=axs[i-1], annot=True)
```

```python
%%time
from xgboost import XGBRegressor
from sklearn.metrics import r2_score
import time

for i in [4]:
    dummy_df=df[i].copy()
    dummy_df2=dummy_df.copy()
    for column in dummy_df.columns:
        start=time.time()

        print("processing column:", column)
        #if isna==0 continue
        if dummy_df[column].isna().sum()==0:
            continue
        col_na_id=dummy_df[dummy_df[column].isnull()].index
        col_train=dummy_df.drop(col_na_id,axis=0)
        col_test=dummy_df[dummy_df.index.isin(col_na_id)]
        X=col_train.drop(column,axis=1)
        y=col_train[column]
        model=XGBRegressor(n_estimators=2000,tree_method='gpu_hist')
        model.fit(X,y)
        score=model.score(X,y)
        print("score:",score)
        pred=model.predict(col_test.drop(column,axis=1))
        dummy_df2[column][col_na_id]=pred
        end=time.time()
        print("time:",end-start)
    df[i]=dummy_df2.copy()
```

```python
# simple imputation on other columns
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
for i in [1,3]:
    df[i][:]=imputer.fit_transform(df[i])
```

```python
from tqdm import tqdm
merged=pd.concat([df[1],df[2],df[3],df[4]],axis=1)
submission = pd.read_csv('data/sample_submission.csv',index_col='row-col')
for s in tqdm(submission.index):
    row,col=s.split('-')
    submission.loc[s,'value'] = merged.loc[int(row),col]
#dtype to double
submission['value']=submission['value'].astype(float)
submission.to_csv('data/submission_gbm.csv')
```

```python
#keras model
import tensorflow as tf
# 80->40->20->1
def my_model():
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.Dense(20, input_shape=(14,), activation='relu'))

    model.add(tf.keras.layers.Dense(20, activation='swish'))
    model.add(tf.keras.layers.Dense(1,activation = 'linear'))
    model.compile(loss='mse', optimizer='adam')
    return model
#callbacks
early_stop = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=3)
reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(monitor='loss', factor=0.1, patience=2)
epoch=50
batch_size=4096
```

```python
%%time
import seaborn as sns
features = list(data.columns)
features_1, features_2, features_3, features_4 = [], [], [], []
F = [[], [], [], [], []]
for feature in features:
    for i in [1, 2, 3, 4]:
        if feature.split('_')[1] == str(i):
            F[i].append(feature)
df = [[], [], [], [], []]

fig, axs = plt.subplots(nrows=4, ncols=1, figsize=(18, 30))

for i in [1, 2, 3, 4]:
    df[i] = data[F[i]]
    corr = df[i].corr()
    sns.heatmap(corr, ax=axs[i-1], annot=True)
```

```python
%%time
import time
#standard scalar
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
for i in [4]:
    dummy_df=df[i].copy()
    dummy_df2=dummy_df.copy()
    for column in dummy_df.columns:
        start=time.time()
        print("processing column:", column)
        #if isna==0 continue
        if dummy_df[column].isna().sum()==0:
            continue
        col_na_id=dummy_df[dummy_df[column].isnull()].index
        col_train=dummy_df.drop(col_na_id,axis=0)
        col_test=dummy_df[dummy_df.index.isin(col_na_id)]
        X=col_train.drop(column,axis=1)
        y=col_train[column]

        X=scaler.fit_transform(X)
      
        model=my_model()
        model.fit(X,y,epochs=epoch,batch_size=batch_size,callbacks=[early_stop,reduce_lr],verbose=2)
        test=col_test.drop(column,axis=1)
        test=scaler.fit_transform(test)
        pred=model.predict(test)
        pred=pred.reshape(-1)
        dummy_df2[column][col_na_id]=pred
        end=time.time()
        print("time:",end-start)
    df[i]=dummy_df2.copy()
```

```python
# simple imputation on other columns
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
for i in [1,3]:
    df[i][:]=imputer.fit_transform(df[i])
```

```python
from tqdm import tqdm
merged=pd.concat([df[1],df[2],df[3],df[4]],axis=1)
submission = pd.read_csv('data/sample_submission.csv',index_col='row-col')
for s in tqdm(submission.index):
    row,col=s.split('-')
    submission.loc[s,'value'] = merged.loc[int(row),col]
#dtype to double
submission['value']=submission['value'].astype(float)
submission.to_csv('data/submission_nn.csv')
```
