```python
def log_rmse(y_true,y_pred):
    mse= np.sum((np.log(y_pred+1)-np.log(y_true+1))**2)/len(y_true)
    return np.sqrt(mse)

```

```python
# adding time infornmation
data=train.copy()
# concatenate test 
data=pd.concat([data,test],ignore_index=True)
data.shape
target_columns = ['target_carbon_monoxide','target_benzene','target_nitrogen_oxides']
# data[target_columns]=np.log(data[target_columns])

#day of the month
data['day_of_the_month']=data['date_time'].dt.day
#weekday
data['weekday']=data['date_time'].dt.dayofweek
#hour
data['hour']=data['date_time'].dt.hour
#weekend
data['weekend']=data['date_time'].dt.dayofweek.isin([5,6])
sensor_columns = ['sensor_1', 'sensor_2', 'sensor_3', 'sensor_4', 'sensor_5']
weather_columns = ['deg_C', 'relative_humidity', 'absolute_humidity']
# lag

lags=[1,4,12,24]
for col in sensor_columns:
    for lag in lags:
        data[col+'_lag_'+str(lag)]=data[col].shift(lag)
for col in weather_columns:
    for lag in lags:
        data[col+'_lag_'+str(lag)]=data[col].shift(lag)
    
```

```python
data.drop(['date_time'],axis=1,inplace=True)
train_data=data.iloc[:len(train)]
test_data=data.iloc[len(train):]
y1 = train_data['target_carbon_monoxide']
y2 = train_data['target_benzene']
y3=train_data['target_nitrogen_oxides']
X=train_data.drop(['target_carbon_monoxide','target_benzene','target_nitrogen_oxides'],axis=1)
test_data=test_data.drop(['target_carbon_monoxide','target_benzene','target_nitrogen_oxides'],axis=1)
y=pd.concat([y1,y2,y3],axis=1)
y.columns=['target_carbon_monoxide','target_benzene','target_nitrogen_oxides']
split=0.05
```

```python
# lightgbm
import lightgbm as lgb
from sklearn.model_selection import TimeSeriesSplit
tscv=TimeSeriesSplit(n_splits=5)
lgb_params = {
    'boosting_type': 'gbdt',
    'objective': 'regression',
    'metric': 'rmse',
    'learning_rate': 0.05,
    'verbose': 0,
    
}
error=[0,0,0]
pred_test=np.zeros((test_data.shape[0],3))
tscv=TimeSeriesSplit(n_splits=5)

for train_index, test_index in tscv.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y1.iloc[train_index], y1.iloc[test_index]
    lgbmReg = lgb.LGBMRegressor(**lgb_params)

    lgbmReg.fit(X_train, y_train, verbose=False)
    y_pred=lgbmReg.predict(X_test)
    error[0]+=log_rmse(y_test,y_pred)
    
    pred_test[:,0]+=lgbmReg.predict(test_data)

error[0]=error[0]/5
print("LogRMSE: %f" % (error[0]))

```

```python
## NN nodel
import tensorflow as tf
# concatenate all ys

# dense with batch normalization
model=tf.keras.Sequential()
model.add(tf.keras.layers.Dense(100,activation='selu',input_shape=(X.shape[1],)))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Dense(100,activation='selu'))
model.add(tf.keras.layers.BatchNormalization())
model.add(tf.keras.layers.Dense(100,activation='selu'))
model.add(tf.keras.layers.BatchNormalization())
# output layer
model.add(tf.keras.layers.Dense(3,activation='linear'))
msle = tf.keras.losses.MeanSquaredLogarithmicError()

model.compile(optimizer='adam',loss=msle)
reduce_lr=tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss',factor=0.5,patience=5,min_lr=0.00001)
early_stop=tf.keras.callbacks.EarlyStopping(monitor='val_loss',patience=10,min_delta=0.00001)
model.fit(X_train,y_train,epochs=100,batch_size=128,validation_data=(X_test,y_test),callbacks=[reduce_lr,early_stop],verbose=2)
```
